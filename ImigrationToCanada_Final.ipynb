```python
#Step 2:
#Loading Canadian Immigration Data into a DataFrame(Importing CSV Data from a URL into Pandas)
df_canada=pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.csv')
print('Data read into a pandas dataframe!')

#Step 3:
#Previewing the First 5 Rows of the Dataset
df_canada.head()
```

```python
# *******Data wrangling*********
#Step 4:
#Check for missing or null values
df_canada.isnull().sum()
```

```python
# Step 5:
#Dataset Summary and Structure Overview
df_canada.info()
```

```python
#Step 6:Listing All Column Names in the DataFrame
df_canada.columns
```

```python
#Step 7:Setting 'Country' as the DataFrame Index
df_canada.set_index('Country',inplace=True)
df_canada
```

```python
#Step 8:
#Creating a List of Year Columns.
years=list(map(str,range(1980,2014)))
years
```

```python
#******* Exploratory Data Analysis (EDA)*******
# Step 9: Getting summary statistics (mean, count, std, min, max, etc.) for numerical colum
df_canada.describe()
```

```python
#Step 10: 
# 10.1: Viewing Available Matplotlib Styles and Setting ggplot Theme.
print(plt.style.available)

#10.2: Applying ggplot Style to All Plots
mpl.style.use(['ggplot'])
```

```python
#Step 11: now going to focus on "Haiti" and visualize how immigration from Haiti to Canada 
#changed over time, especially around 2010 after the earthquake.


#11.1: Checking if 'Haiti' Exists in the Dataset Index
"Haiti" in df_canada.index
```

```python
#11.2:Extracting Haitiâ€™s Immigration Data for 1980â€“2013.
haiti=df_canada.loc['Haiti', years]
haiti.head()
```

```python
#11.3: Annotated Line Plot of Immigration from Haiti (1980â€“2013)
haiti.index = haiti.index.map(int) 
haiti.plot(kind='line')

plt.title('Immigration from Haiti to Canada (1980â€“2013)')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

# annotate the 2010 Earthquake. 
# syntax: plt.text(x, y, label)
plt.text(2000, 6000, '2010 Earthquake',color='blue') 

plt.show() 
```

```python
#Step 12: Let's compare the number of immigrants from India and China from 1980 to 2013.
#Step 12.1:Checking if 'China'  Exists in the Dataset Index
"China" in df_canada.index
```

```python
#Step 12.2: Checking if 'India'  Exists in the Dataset Index
"India" in df_canada.index
```

```python
#Step 12.3: Extracting Immigration Data for India and China (1980â€“2013).
df_CI=df_canada.loc[['China','India'],years]
df_CI
```

```python
#12.4: Transposing Indiaâ€“China Dataset for Year-Based Indexing.
df_CI=df_CI.transpose()
df_CI.head()
```

```python
#12.5:Line Plot Comparing Immigration from China and India (1980â€“2013)

#Converting Year Index to Integer Format
df_CI.index=df_CI.index.map(int)
df_CI.plot(kind='line')

plt.title('Immigrants from China and India')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

```python
#Step 13: Line Plot of Immigration Trends for the Top 5 Countries (1980â€“2013).

#13.1: Sorting Countries by Total Immigration in Descending Order.
df_canada.sort_values(by='Total', axis=0, ascending=False, inplace=True)

#13.2: Selecting the Top 5 Countries by Immigration.
df_top5=df_canada.head(5)
df_top5
```

```python
#13.3: Keep only year columns and transpose.
df_top5=df_top5[years].transpose()
df_top5
```

```python
#13.4: Converting Year Index to Integer for Plotting.
df_top5.index=df_top5.index.map(int)

#13.5: Line Plot of Immigration Trends for Top 5 Countries.
df_top5.plot(kind='line', figsize=(14, 8))
plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()
```
